# SQL TASK










### Create Table


```python
import sqlite3

conn = sqlite3.connect("FirewallAudit.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS Firewall_Logs (
    Log_ID INTEGER PRIMARY KEY,
    Traffic_Type TEXT,
    Bandwidth_KB INTEGER,
    Firewall_Action TEXT
)
""")
conn.commit()
print("Firewall Table Created Successfully.")
```

### Insert Records

```python
cursor.executemany("""
INSERT INTO Firewall_Logs (Log_ID, Traffic_Type, Bandwidth_KB, Firewall_Action)
VALUES (?, ?, ?, ?)
""", [
    (501, "Web-HTTP", 1200, "Allow"),
    (502, "File-FTP", 980000, "Drop"),
    (503, "Email-SMTP", 4500, "Allow"),
    (504, "Unknown-P2P", 750000, "Block"),
    (505, "Secure-HTTPS", 25000, "Allow")
])
conn.commit()
print("5 Firewall Records Inserted Successfully.")
```

Display All Records

```python
cursor.execute("SELECT * FROM Firewall_Logs")

for row in cursor.fetchall():
    print(row)
```

### WHERE Clause (Filtering Data)

```python
cursor.execute("""
SELECT * FROM Firewall_Logs
WHERE Bandwidth_KB > 50000
""")

for row in cursor.fetchall():
    print(row)
```

### UPDATE Statement

```python
cursor.execute("""
UPDATE Firewall_Logs
SET Firewall_Action = 'Allow'
WHERE Log_ID = 504
""")
conn.commit()
print("Firewall Action Updated Successfully.")
```

### DELETE Statement

```python
cursor.execute("""
DELETE FROM Firewall_Logs
WHERE Log_ID = 502
""")
conn.commit()
print("High-Risk Record Deleted Successfully.")
```

### Aggregate Functions (SUM)

```python
cursor.execute("""
SELECT SUM(Bandwidth_KB) FROM Firewall_Logs
""")

print("Total Network Bandwidth (KB):", cursor.fetchone()[0])
```

### GROUP BY Clause

```python
cursor.execute("""
SELECT Firewall_Action, COUNT(*)
FROM Firewall_Logs
GROUP BY Firewall_Action
""")

for row in cursor.fetchall():
    print(row)
```

### Terminate Session

```python
conn.close()
print("Database Connection Closed Cleanly.")
```
