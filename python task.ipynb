 Load Data

```python
import pandas as pd
import numpy as np

data = {
    "Server_ID": ["S001", "S002", "S003", "S004", "S005", "S006", "S007", "S008"],
    "CPU_Usage_Pct": [12.5, 45.2, 98.1, 15.4, 18.5, 99.4, 22.1, 14.8],
    "Memory_Usage_Pct": [30.2, 45.8, 95.4, 33.1, 35.0, 97.2, 40.5, 31.9],
    "Threat_Label": ["Normal", "Normal", "Attack", "Normal", "Normal", "Attack", "Normal", "Normal"]
}

df = pd.DataFrame(data)
print(df.head())
```

Check Missing Values

```python

print(df.isnull().sum())
```

WHERE Clause (Filtering Data in Python)

```python
high_load = df[df["CPU_Usage_Pct"] > 80]
print(high_load)
```

Bar Chart Graph

```python
import matplotlib.pyplot as plt

# Data for the bar graph
threats = ["DDoS", "Malware", "Phishing"]
counts = [45, 80, 25]

plt.figure(figsize=(6, 4))
plt.bar(threats, counts, color=["red", "blue", "green"])

plt.title("Total Threats Detected")
plt.xlabel("Threat Name")
plt.ylabel("Number of Attacks")

plt.show()
```

Loops

```python
print("--- For Loop ---")
for count in range(3):
    print("Scanning system number:", count)

print("--- While Loop ---")
x = 1
while x <= 2:
    print("Loop running step:", x)
    x = x + 1
```
