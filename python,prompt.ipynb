```python
Prompt 2 : Apply One-Hot Encoding to Protocol Values
```

```python
import pandas as pd
import numpy as np

raw_data = {
    "Duration_Sec": [12.5, 450.2, 3.1, 89.4, 1200.5, 15.2, 45.8, 2200.1],
    "Protocol_Type": ["TCP", "UDP", "TCP", "ICMP", "TCP", "UDP", "TCP", "TCP"],
    "Packets_Sent": np.random.randint(10, 1000, 8).tolist(),
    "Bytes_Downloaded": [1024, 512000, 256, 1024, 85000000, 2048, 4096, 120000000]
}
df = pd.DataFrame(raw_data)

# Transform text category column into multiple binary flags
df_encoded = pd.get_dummies(df, columns=["Protocol_Type"], dtype=int)
print(df_encoded.head(4))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np

raw_data = {
    "Packets_Sent": np.random.randint(10, 1000, 8).tolist(),
    "Bytes_Downloaded": [1024, 512000, 256, 1024, 85000000, 2048, 4096, 120000000]
}
df = pd.DataFrame(raw_data)

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df)

df_scaled = pd.DataFrame(scaled_features, columns=["Scaled_Packets", "Scaled_Bytes"])
print(df_scaled.head(4))
```

```python
Prompt 5 : Compute Mathematical Security Risk Matrix Scores
```

```python
import pandas as pd
import numpy as np

raw_data = {
    "Threat_Probability": [0.12, 0.88, 0.05, 0.95, 0.45],
    "Asset_Criticality_Weight": np.random.randint(1, 6, 5).tolist(), # Scale from 1 to 5
    "Data_Volume_MB": [5, 1200, 1, 15000, 450]
}
df = pd.DataFrame(raw_data);

# Log scaling normalizes huge spikes in downloaded MB data ranges
df["Volume_Score"] = np.log1p(df["Data_Volume_MB"]);
df["Volume_Score_Normalized"] = (df["Volume_Score"] - df["Volume_Score"].min()) / (df["Volume_Score"].max() - df["Volume_Score"].min());

# Calculate composite risk calculation formula
df["Calculated_Risk_Score"] = (df["Threat_Probability"] * 0.5 + df["Asset_Criticality_Weight"]/5 * 0.3 + df["Volume_Score_Normalized"] * 0.2) * 100;
df["Calculated_Risk_Score"] = df["Calculated_Risk_Score"].round(1);

print(df[["Threat_Probability", "Asset_Criticality_Weight", "Calculated_Risk_Score"]]);
```
